```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

client_ids = [33615, 66037, 46820, 38378, 24741]

# Filter DataFrame for only the specified clients
filtered_df = df[df['client_id'].isin(client_ids)]

# Organize the required info by client_id
summary = filtered_df.groupby('client_id').agg({
    'qty': 'sum',                        # Total units purchased
    'shipping_price': 'sum',             # Total shipping price
    'line_price': 'sum',                 # Total revenue
    'line_cost': 'sum',                  # Total costs
    'line_profit': 'sum'                 # Total profit
}).reset_index()

# Sort the DataFrame by total profit in descending order
sorted_summary = summary.sort_values(by='line_profit', ascending=False)

# Manually set the index to reverse order starting from the number of rows minus 1
sorted_summary.reset_index(drop=True, inplace=True)  # Reset the index to default integer index
sorted_summary.index = sorted_summary.index[::-1]  # Reverse the index

# Print the results with the client_id column first
sorted_summary = sorted_summary[['client_id', 'qty', 'shipping_price', 'line_price', 'line_cost', 'line_profit']]
sorted_summary
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# Convert monetary values to millions and format
summary['shipping_price'] = (sorted_summary['shipping_price'] / 1e6).map("${:.2f}M".format)
summary['line_price'] = (sorted_summary['line_price'] / 1e6).map("${:.2f}M".format)
summary['line_cost'] = (sorted_summary['line_cost'] / 1e6).map("${:.2f}M".format)
summary['line_profit'] = (sorted_summary['line_profit'] / 1e6).map("${:.2f}M".format)

# Rename columns for presentation
summary.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Units',
   'shipping_price': 'Shipping',
    'line_price': 'Total Revenue',
    'line_cost': 'Total Cost',
    'line_profit': 'Total Profit'
}, inplace=True)

# Display the formatted DataFrame to verify changes
print(summary.head())
```

```python
# Sort the updated data by "Total Profit" form highest to lowest

summary.sort_values(by='Total Profit', ascending=False, inplace=True)
summary.reset_index(drop=True, inplace=True)  
summary.index = summary.index[::-1]  
summary
```
